```python
import pandas as pd

data = pd.read_csv("C:\\Users\\User\\CIS437L_dataset1.csv")
df = pd.DataFrame(data)

print(df.info())
```

```python
data = pd.read_csv("C:\\Users\\User\\CIS437L_dataset4.csv")
df4 = pd.DataFrame(data)

print(df4.info())
```

```python
#Element 1 Artifact 1
import numpy as np
import statistics

a = df['Monthly (%)']

monthly = np.array(a)

Average = statistics.mean(monthly)
print("The average monthly % from May 2019 to May 2023:", round(np.mean(Average), 2))
```

```python
#Element 1 Artifact 2
index, Month = "Index", "Monthly (%)"

correlation = df[index].corr(df[Month])
print("Correlation between the Index and Monthly %", round(correlation, 2))
```

```python
#Element 2 Artifact 1
genre = df4['Genre'].value_counts()

print(genre)
```

```python
#Element 2 Artifact 2
publisher = df4['Publisher'].value_counts()

print(publisher.head())
```

```python
#Element 3 Artifact 2
NA, JP = "NA_Sales", "JP_Sales"

correlation = df4[NA].corr(df4[JP])
print("Correlation between the North American Sales and Japanese Sales:", round(correlation, 2))
```

```python
#Element 3 Artifact 1
b = df4['Global_Sales']

sales = np.array(b)

Average = statistics.mean(sales)
print("The average Global Sales Sales for video games:", round(np.mean(Average), 2))
```
