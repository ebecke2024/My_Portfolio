```python
import pandas as pd

data1 = pd.read_csv("C:\\Users\\User\\anime_dvd.csv")
df1 = pd.DataFrame(data1)

data2 = pd.read_csv("C:\\Users\\User\\apparel.csv")
df2 = pd.DataFrame(data2)

data3 = pd.read_csv("C:\\Users\\User\\manga.csv")
df3 = pd.DataFrame(data3)

data4 = pd.read_csv("C:\\Users\\User\\novel.csv")
df4 = pd.DataFrame(data4)
```

```python
#Counting the number Genres we have in stock in manga
Genre = df3['genre'].value_counts()

print(Genre)
```

```python
#Counting the number Genres we have in stock in novels
Genre2 = df4['genre'].value_counts()

print(Genre2)
```

```python
#Finding a correlation between manga and apparel sales
manga_sales = "Manga_Qty_Sold"
Apparel_sales = "App_Qty_Sold"

correlation = df3[manga_sales].corr(df2[Apparel_sales])
print("The correlation between manga and apparel sales is ", round(correlation, 2))
```

```python
#Finding a correlation between manga and novel sales
manga_sales = "Manga_Qty_Sold"
Novel_sales = "Novel_Qty_Sold"

correlation = df3[manga_sales].corr(df4[Novel_sales])
print("The correlation between manga and Novel sales is ", round(correlation, 2))
```

```python
#Finding a correlation between Apparel and Anime sales
Apparel_sales = "App_Qty_Sold"
Anime_Sales = "Anime_Qty_Sold"

correlation = df2[Apparel_sales].corr(df1[Anime_Sales])
print("The correlation between Anime and apparel sales is ", round(correlation, 2))
```

```python
#Counting the type of dvds we have 
Type = df1['Type'].value_counts()

print(Type)
```
